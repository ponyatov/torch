```python
# класс направленных графов
from graphviz import Digraph
```

```python
# пустой граф
dot = Digraph() ; dot
```

```python
# два узла
dot.node('1')
dot.node('2')
# список ребер
dot.edges([['1','2']])
# для односимвольных имен узлов можно использовать итерацию по строке
dot.edges(('12',))
```
